```python
# import libraries
import pandas as pd
import numpy as np

# read in csv file
df = pd.read_csv('netflix_titles_nov_2019.csv')
```

```python
# summarise dataframe columns
df.info()
```

```python
# convert date_added to datetime format
df['date_added'] = pd.to_datetime(df['date_added'])

df.info()
```

```python
# add new column showing month-year for date_added
df['year_month'] = df['date_added'].dt.to_period('M')

df.head()
```

```python
# convert date_added from datetime to date format
df['date_added'] = df['date_added'].dt.date

df.head()
```

```python
# replace missing values in director column with 'unknown'
df['director'].fillna('unknown', inplace=True)

df.info()
```

```python
# replace missing values in cast column with 'unknown'
df['cast'].fillna('unknown', inplace=True)

df.info()
```

```python
# replace missing values in country column with 'unknown'
df['country'].fillna('unknown', inplace=True)

df.info()
```

```python
# replace missing values in rating column with 'no rating'
df['rating'].fillna('no rating', inplace=True)

df.info()
```

```python
# check if any rows are duplicated
duplicates = df.duplicated()

df[duplicates]
```

```python
# save cleaned dataframe as a csv
df.to_csv('netflix_titles_nov_2019_cleaned.csv')
```
